```python
import pandas as pd
```

```python
df = pd.read_csv('transaction-fees.csv', names=['date', 'fee_usd'])
df_ = pd.read_csv('n-transactions.csv', names=['date', 'tx_num'])
```

```python
df_merge = pd.merge(df, df_, on='date', how='left')
df_merge = df_merge.rename(columns={'fee_usd': 'fee_btc'})
```

```python
df_merge
```

```python
df_merge['btc_per_tx'] = df_merge.fee_btc / df_merge.tx_num
df_merge.date = pd.to_datetime(df_merge.date)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize=(15, 10))
df_merge.plot.line('date', 'btc_per_tx')
plt.ylim(0, 0.0025)
```

```python
df_merge.plot.line('date', 'btc_per_tx', logy=True)
plt.xlim('2011-01-01', '2018-01-01')
```
